```python
import time
import os
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.by import By
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.chrome.options import Options
```

```python
Path = "chromedriver.exe"
url = "https://firebase.google.com/codelabs/contentrecommendation-android?hl=zh-tw#0"
s = Service(Path)
driver = webdriver.Chrome(service=s)
driver.get(url)
search = driver.find_element("class","codelab-title")
print(search.text)
driver.quit()
```

設定driver執行檔路徑

```python
options = Options()
options.chrome_executable_path = "chromedriver.exe"
```

```python
driver = webdriver.Chrome(options=options) #建立driver物件
driver.get("https://www.ptt.cc/bbs/movie/index.html") #連接ppt電影版
# print(driver.page_source) 網頁的所有程式碼
titles = driver.find_elements(By.CLASS_NAME, "title")
for title in titles:
    print(title.text)
```

# 取得PPT電影版的上一頁

```python
driver = webdriver.Chrome(options=options)
driver.get("https://www.ptt.cc/bbs/movie/index.html")
link = driver.find_element(By.LINK_TEXT, "‹ 上頁") #注意此element沒有s因為他只需要找一個標籤
link.click()
print(driver.page_source)
```

```python
driver = webdriver.Chrome(options=options)
driver.get("https://www.ptt.cc/bbs/movie/index.html")
link = driver.find_element(By.LINK_TEXT, "最舊")
link.click()
tags = driver.find_elements(By.CLASS_NAME, "title")
for tag in tags:
    print(tag.text)
```
